```python
def resize_images(diretorio, largura, altura):
    # Obtém a lista de arquivos no diretório
    arquivos = os.listdir(diretorio)

    for arquivo in arquivos:
        # Verifica se o arquivo é uma imagem
        if arquivo.endswith(('.jpg', '.jpeg', '.png')):
            # Caminho completo do arquivo
            caminho_arquivo = os.path.join(diretorio, arquivo)

            # Carrega a imagem usando o OpenCV
            imagem = cv2.imread(caminho_arquivo)

            # Redimensiona a imagem para a largura e altura desejadas
            imagem_redimensionada = cv2.resize(imagem, (largura, altura))

            # Salva a imagem redimensionada
            cv2.imwrite(caminho_arquivo, imagem_redimensionada)
```
